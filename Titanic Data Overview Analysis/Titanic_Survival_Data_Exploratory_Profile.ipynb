```python
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df
```

```python
from ydata_profiling import ProfileReport
prof= ProfileReport(df,title='Titanic Survival Data – Exploratory Profile')
prof.to_file(output_file='titanic_raw_data_overview.html')
```

```python
import matplotlib.pyplot as plt
sns.kdeplot(df[df['survived']==0]['age'], label='Not Survived')
sns.kdeplot(df[df['survived']==1]['age'], label='Survived')
plt.legend()
plt.show()
```

```python
import pandas as pd
```

```python
cross_tab = pd.crosstab(df['sex'], df['survived'])
display(sns.heatmap(cross_tab))
```

```python
sns.heatmap(pd.crosstab(df['pclass'],df['survived']))
```

This table shows the count of individuals by their sex and survival status. The rows represent 'sex' (female, male) and the columns represent 'survived' (0 for not survived, 1 for survived). This quickly tells us how many females and males survived or did not survive.

```python
df.groupby('pclass')['survived'].mean()
```
